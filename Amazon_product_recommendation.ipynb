```python
from sentence_transformers import SentenceTransformer
from torch.utils.data import TensorDataset, DataLoader
import torch
```

```python
sent_trainformer_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
### Keep a track of product info - note in the df_train, product_id and about_product are not unique, only user_ids
print(f'Train len {len(df_train)}')

product_about_info = df_train[['product_id','about_product']].drop_duplicates()

print(f'product info len {len(product_about_info)}')
```

### CONVERT ABOUT PRODUCT TO EMBEDDINGS

```python
product_texts = product_about_info['about_product'].tolist() ### convert only unique about

embeddings = sent_trainformer_model.encode(product_texts, convert_to_tensor=True)

embeddings.shape
```

```python
### SETUP RQVAE TRAINING
```

```python
dataset = TensorDataset(embeddings)
batch_size = 64
loader  = DataLoader(dataset, batch_size=batch_size, shuffle=True, drop_last=True)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

rqvae_model = RQVAE(
    input_dim=embeddings.size(1),     # 384 for all-MiniLM-L6-v2
    hidden_dims=[512, 256],           # or your own choice
    latent_dim=128,
    num_quantizers=4,
    codebook_size=256,
    commitment_cost=0.25
).to(device)

optimizer = torch.optim.Adam(rqvae_model.parameters(), lr=3e-4, weight_decay=1e-5)
```

```python
num_epochs = 5 #50

for epoch in range(1, num_epochs+1):
    rqvae_model.train()
    running_loss = 0.0

    for (batch_embeddings,) in loader:
        batch_embeddings = batch_embeddings.to(device)

        # 1) Forward pass
        outputs = rqvae_model(batch_embeddings)

        # 2) Compute RQVAE loss
        loss = rqvae_model.loss_function(batch_embeddings, outputs)

        # 3) Backward + step
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * batch_embeddings.size(0)

    epoch_loss = running_loss / len(loader.dataset)
    print(f"Epoch {epoch:02d} — Loss: {epoch_loss:.4f}")
```

```python
### At this point we have learned the codebooks - Now loading codebook codes
```

```python
rqvae_model.eval()
with torch.no_grad():
  _, all_codes = rqvae_model.quantize(rqvae_model.reparameterize(*rqvae_model.encode(embeddings)))
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
import torch.nn.functional as F
```

```python
class SessionDataset(Dataset):
    def __init__(self, sessions):
        self.examples = []
        for sess in sessions:
            if len(sess) < 2: continue
            src = torch.tensor(sess[:-1], dtype=torch.long)
            tgt = torch.tensor(sess[1:],  dtype=torch.long)
            self.examples.append((src, tgt))
    def __len__(self):   return len(self.examples)
    def __getitem__(self, i): return self.examples[i]


def collate_fn(batch, pad_idx=0):
    src_batch, tgt_batch = zip(*batch)
    src_padded = pad_sequence(src_batch, batch_first=True, padding_value=pad_idx)
    tgt_padded = pad_sequence(tgt_batch, batch_first=True, padding_value=pad_idx)
    src_kp = src_padded.eq(pad_idx)
    tgt_kp = tgt_padded.eq(pad_idx)
    return src_padded, tgt_padded, src_kp, tgt_kp
```

```python
dataset = SessionDataset(tokenized_sessions)
dataloader = DataLoader(dataset,
                        batch_size=32,
                        collate_fn=collate_fn)
```

```python
### TRAIN OUR TRANSFORMER MODEL
```

```python
from models.transformer import SemanticIDTransformer
import torch.nn as nn
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SemanticIDTransformer(vocab_size = 1351 ).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
criterion = nn.CrossEntropyLoss(ignore_index=0)
```

```python
#model.forward
```

```python
num_epochs = 2
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for src, tgt, src_kp, tgt_kp in dataloader:
        src, tgt = src.to(device), tgt.to(device)
        src_kp, tgt_kp = src_kp.to(device), tgt_kp.to(device)

        # decoder input / labels
        tgt_input = tgt[:, :-1]
        labels    = tgt[:,  1:]

        # causal mask: (T-1, T-1)
        T = tgt_input.size(1)
        mask = model.transformer.generate_square_subsequent_mask(T).to(device)

        optimizer.zero_grad()
        logits = model(
            src, tgt_input,
            src_key_padding_mask=src_kp,
            tgt_key_padding_mask=tgt_kp[:, :-1],
            tgt_mask=mask
        )
        # print(logits.shape)
        # print(labels.shape)
        loss = criterion(
            logits.reshape(-1, logits.size(-1)),
            labels.reshape(-1)
        )
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1} — avg loss: {total_loss/len(dataloader):.4f}")
```

```python
### This will generate 5*4 tokens whcih will then be split into 4's (codebooks)
### This is greedy implementation (argmax) - We can use beam search instead which promotes
## more diversity.
## They also discuss topk sampling in the paper

def generate_semantic_ids(model, input_seq, num_items=5, tokens_per_item=4):
    model.eval()
    generated = []
    tgt = torch.zeros(1, 1).long().to(input_seq.device)  # BOS token or just 0

    total_tokens = num_items * tokens_per_item

    for _ in range(total_tokens):
        out = model(input_seq, tgt)
        next_token = out[:, -1, :].argmax(dim=-1, keepdim=True)
        tgt = torch.cat([tgt, next_token], dim=1)
        generated.append(next_token.item())

    # Skip the initial 0 and divide into item chunks
    print(generated)
    #generated = generated[1:]
    semantic_ids = [generated[i:i+tokens_per_item] for i in range(0, len(generated), tokens_per_item)]
    return semantic_ids
```
